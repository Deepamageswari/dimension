```python
import pandas as pd
import math
input_date=input("enter the date in the format- 2021-01-01")
pd.set_option('display.max_rows',None)
from datetime import datetime, date
s=pd.date_range(input_date,periods=365,freq='D')
df=pd.DataFrame(s,columns=['Date'])

df=pd.DataFrame(s,columns=['Date'])
df.insert(0,'Dates',s)
df.insert(1,'year',df["Date"].dt.strftime("%y"))
df.insert(2,'month',df["Date"].dt.strftime("%m"))
df.insert(3,'yearmonth',df["Date"].dt.strftime("%y/%m"))#primary key should be
df.insert(4,'monthnm',df['Date'].dt.strftime("%B"))
df.insert(5,'monthabb',df['Date'].dt.strftime("%b"))
df.insert(6,'day',df['Date'].dt.strftime("%j"))
df.insert(7,'day_week',(df['Date'].dt.dayofweek)) 
df.insert(8,'day_nm_week',(df['Date'].dt.day_name())) #df.insert(8,'day_week_abbr',df['Date'].dt.strftime("%A"))
df.insert(9,'day_week_abbr',df['Date'].dt.strftime("%a"))
df.insert(10,'quarter',df['Date'].dt.quarter)#df['Date'].dt.quarter
df.insert(11,'season','none')
df.insert(12,'Holiday','none')
today=datetime.now()
df.insert(13,'current_year',df['year'].apply(lambda x:'Y' if x==today.strftime("%y") else 'N'))#.dt.strftime("%y"))
df.insert(14,'prioryear',df['year'].apply(lambda x:'Y' if x==str((int(today.strftime("%y"))-1)) else 'N'))#df['<= 53'] = df['mynumbers'].apply(lambda x: 'True' if x <= 53 else 'False')
df.insert(15,'current_month',df['month'].apply(lambda x:'Y' if x==today.strftime("%m") else 'N'))
df.insert(16,'prior_month',df['month'].apply(lambda x:'Y' if x==str((int(today.strftime("%m"))-1)).zfill(2) else 'N'))
df.insert(17,'first_of_month',df['Date'].dt.strftime('%d').apply(lambda x:'Y' if x=='01' else 'N'))
df.insert(18,'business_day',"N")
df.loc[(df['day_nm_week']!='Saturday')&(df['day_nm_week']!='Sunday')&(df['Date']!='2021-01-01')\
       &(df['Date']!='2021-01-16')&(df['Date']!='2021-01-26')&(df['Date']!='2021-02-18')&(df['Date']!='2021-05-01')\
                                                                                       &(df['Date']!='2021-08-15')\
       &(df['Date']!='2021-11-01')&(df['Date']!='2021-12-31'),'business_day']='B'
df.insert(15,'fiscal_month',df['month'])
df.loc[(df['monthnm']=="January")|(df['monthnm']=="February")|(df['monthnm']=="December"),'season']= 'winter'
df.loc[(df['monthnm']=="September")|(df['monthnm']=="October")|(df['monthnm']=="November"),'season']= 'autumn'
df.loc[(df['monthnm']=="June")|(df['monthnm']=="July")|(df['monthnm']=="August"),'season']= 'summer'
df.loc[(df['monthnm']=="March")|(df['monthnm']=="April")|(df['monthnm']=="May"),'season']= 'spring'


df.loc[df['Date'] == '2021-01-01', 'Holiday'] = 'New Year'
df.loc[df['Date'] == '2021-01-16', 'Holiday'] = 'Shankaranthi'
df.loc[df['Date'] == '2021-01-26', 'Holiday'] = 'Republic Day'
df.loc[df['Date'] == '2021-02-18', 'Holiday'] = 'ID Day'
df.loc[df['Date'] == '2021-05-01', 'Holiday'] = 'May Day'
df.loc[df['Date'] == '2021-08-15', 'Holiday'] = 'Independence Day'
df.loc[df['Date'] == '2021-11-01', 'Holiday'] = 'Karanataka Day'
df.loc[df['Date'] == '2021-12-31', 'Holiday'] = 'Last day of the Year'
df.set_index('Date',inplace=True)
#df.loc[df['days'] == 'Saturday', 'Holiday'] = 'One shift'
#df.loc[df['days'] == 'Sunday', 'Holiday'] = 'Normal Holiday'
print(df)
```
